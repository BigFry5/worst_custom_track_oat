```python
import pandas as pd
import numpy as np
from google.colab import files
```

```python
my_file = files.upload()
data = pd.read_csv('custom_tracks.csv')
```

```python
df = pd.DataFrame(data)
df
```

```python
titles = df['Title']
name_lengths = [len(title) for title in titles]
df['Name Length'] = name_lengths
df
```

```python
length_penalty = []
X = df['Name Length']
X = (X-np.mean(X))/np.std(X)
for i in range(len(titles)):

   for j in titles[i]:
     count = 0
     if j == ' ':
       count += 1

   length_penalty.append(count*0.05 + 0.09*(len(titles[i])))

df['length penalty'] = length_penalty
X = abs(X)-df['length penalty']
df['weird-length score'] = X
df
```

```python
def calculate_non_alpha_score(df):

    scores = []
    for j in df:
        count = 0
        for i in j.replace(' ',''):
          if not i.isalpha():
            count += 1
        scores.append(count/len(j))
    return scores
```

```python
df['non_alpha_score'] = calculate_non_alpha_score(df["Title"])
df
```

```python
total_scores = df['weird-length score'] + df['non_alpha_score']*1.68
df['total score'] = total_scores
df.sort_values(by='total score', ascending=False)
```
